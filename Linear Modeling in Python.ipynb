# How well does the size of a condominium in New York City explain sale price?

In this project we'll explore how well the size of a condominium (measured in gross square feet) explains, or predicts, sale price in New York City. We will also explore how well the size of a condominium predicts sale price in each of the five boroughs of New York City: the Bronx, Brooklyn, Manhattan, Staten Island, and Queens.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in dataset
NYC_property_sales = pd.read_csv('nyc_property_sales.csv')

# Print out the first rows of the dataset
NYC_property_sales.head()
```

```python
# Print out the dimension of the dataset
NYC_property_sales.shape
```

# Filtering the building class ("R4")

For this project we will only work with a single type of building class ("R4"), a condominium residential unit in a building with an elevator. This building class is the most common building class in this `NYC_property_sales` dataframe. 

```python
# Filter to include only property type: CONDO; RESIDENTIAL UNIT IN ELEVATOR BLDG.
NYC_condos = NYC_property_sales.loc[NYC_property_sales['building_class_at_time_of_sale'] == 'R4']
```

```python
# Print out the first rows of the dataset
NYC_condos.head()
```

```python
# Print out the dimension of the dataset
NYC_condos.shape
```

```python
NYC_condos.describe()
```

# Calculating average sale price

```python
# Calculate the average price and round to nearest 2 decimal places, avg_price
avg_sale_price = round(NYC_condos["sale_price"].mean(),2)

# Print the average price
print("The average sale price of Condos - Elevator Apartment in NYC is ${}.".format(avg_sale_price))
```

# Analyzing sale prices by NYC borough

```python
# Group by borough and calculate summary statistics
boroughs = NYC_condos.groupby("borough")["sale_price"].agg(["sum", "mean", "median", "count"])

# Round boroughs to 2 decimal places, and sort by mean in descending order
boroughs = boroughs.round(2).sort_values("mean", ascending=False)

# Print boroughs
print(boroughs)
```

# Outliers and Data Integrity Issues

```python
# We begin our investigation of outliers by sorting all sale records by sale price, from high to low
NYC_condos.sort_values(by='sale_price', ascending=False).head(50)
```

```python
# Create a copy of the original DataFrame
NYC_condos_original = NYC_condos.copy()
```

```python
# Remove 165 East 66th Street sale record
NYC_condos = NYC_condos[NYC_condos['address'] != '165 East 66th St, Resi']
```

# How well does gross square feet explain sale price for all records combined?

We'll take a look at the highest sale price observations in Brooklyn. There are a number of sale records at a sale price of around 30 Million, but there is only a single observations in the range of 10 to 30 Million.

```python
# Filter records for borough "Brooklyn"
brooklyn_condos = NYC_condos[NYC_condos['borough'] == 'Brooklyn']
```

```python
# Sorting all Brooklyn sale records by sale price, from high to low
brooklyn_condos.sort_values(by='sale_price', ascending=False).head(50)
```

Looking through the results we see that there are approximately 40 sales records with a price of 29,620,207. This price point appears to be unusual for Brooklyn. Scrolling through the results we also see that all 40 property sales took place on the same day, 2019-04-08. This indicates that a transaction took place on this date where all 40 units were purchased for a TOTAL price of 29,620,207, 29,620,207 per unit. For our purposes we will remove all 40 observations from the dataset because sale prices for each unit are erroneous

```python
# Group by sale_price and sale_date, filter records with count <= 2, and ungroup
NYC_condos = NYC_condos.groupby(['sale_price', 'sale_date']).filter(lambda x: len(x) <= 2).reset_index(drop=True)
```

```python
# Print out the dimension of the dataset
NYC_condos.shape
```

# Linear Regression Model for Boroughs in New York City Combine

Now that we've removed many multi-unit sales from the dataset, let's generate a linear regression model for all New York City neighborhoods combined. As a reminder, we are predicting sale_price on the basis of gross_square_feet

```python
import statsmodels.api as sm

# Fit a linear regression model
X = sm.add_constant(NYC_condos['gross_square_feet'])
y = NYC_condos['sale_price']
model = sm.OLS(y, X).fit()

# Print the summary of the linear regression model
print(model.summary())
```

```python
# How does this compare to the 'NYC_condos_original' dataframe that includes multi-unit sales?

# Fit a linear regression model
X = sm.add_constant(NYC_condos_original['gross_square_feet'])
y = NYC_condos_original['sale_price']
model = sm.OLS(y, X).fit()

# Print the summary of the linear regression model
print(model.summary())
```

# Comparison of linear modeling results

A bivariate linear regression of 'sale_price' (price) explained by 'gross_square_feet' (size) was performed on two different datasets containing condominium sale records for New York City. One dataset, 'NYC_condos', was cleaned to remove multi-unit sale records (where the same sale price is recorded for many units). The other dataset, 'NYC_condos_original', remained unaltered and contained all original sale records. In each case, the hypothesis is that  there is a relationship between the size of a condominium ('gross_square_feet') and the price ('sale_price'). We can declare there is a relationship between condominium size and price when the slope is sufficiently far from zero. 

For each model, the t-statistic was high enough, and the p-value was low enough, to declare that there is, in fact, a relationship between 'gross_square_feet' and 'sale_price'. The t-statistic for the cleaned dataset (NYC_condos) was nearly double that of the original dataset (NYC_condos_original) at 113.04 versus 61.39. In each case the p-value was well below the 0.05 cutoff for significance meaning that it is extremely unlikely that the relationship between condominium size and sale price is due to random chance. 

The confidence interval for the slope is [4384.254, 4538.999] for the NYC_condos dataset compared to only [1154.636, 1230.802] for the NYC_condos_original dataset. This difference can likely be attributed to the removal of many multi-million dollar sale records for smaller units which impacted price predictions in the original dataset. The measure for *lack of fit*, or residual standard error (RSE) was lower for the cleaned dataset at 2,945,000 compared to 4,745,000 for the original dataset. However, it must be noted that the NYC_condos is smaller than the NYC_condos_original by 150 observations. Finally, the R-squared, or the proportion of the variability in sale_price that can be explained by gross_square_feet is 0.6166 for the cleaned NYC_condos. This is nearly double the R-squared value estimated for the NYC_condos_original dataset at 0.3177. 

Below is the updated scatterplot that uses the cleaned NYC_condos data. For the Brooklyn borough we are better able to see the spread of the data and how the trend line fits the data because we removed the 30 million (USD) outliers. The same is true for the Manhattan borough because the 200 million (USD) multi-unit sale was removed.

```python
# Updated scatterplot 
(ggplot(NYC_condos, aes(x='gross_square_feet', y='sale_price'))
 + geom_point(alpha=0.3)
 + facet_wrap('~ borough', scales='free', ncol=2)
 + geom_smooth(method = "lm")
 + theme_minimal()
 + labs(title='Condominium Sale Price in NYC Generally Increases with Size',
        x='Size (Gross Square Feet)',
        y='Sale Price (USD)')
)
```

# Linear Regression Models for each Borough - Coefficient Estimates

```python
# Group by borough and nest the data
NYC_nested = NYC_condos.groupby('borough').apply(lambda x: x[['gross_square_feet', 'sale_price']].to_dict(orient='records')).reset_index(name='data')

# Print the nested DataFrame
print(NYC_nested)

print("\n")

# Print the first few records of the nested data for the third borough (index 2)
print(NYC_nested['data'][2][:5])
```

```python
# Fit linear models to each borough, individually
from statsmodels.formula.api import ols

linear_models = {}
for i, row in NYC_nested.iterrows():
    borough = row['borough']
    data = pd.DataFrame(row['data'])
    model = ols(formula='sale_price ~ gross_square_feet', data=data).fit()
    linear_models[borough] = model
```

```python
# Print the coefficients for each borough
for borough, model in linear_models.items():
    print(f'Borough: {borough}')
    print(model.summary())
    print('\n')
```

```python
# Extract coefficients and confidence intervals for each borough
coefficients = {}
for borough, model in linear_models.items():
    coef_df = model.params.to_frame(name='Coefficient')
    coef_df['Lower CI'] = model.conf_int()[0]
    coef_df['Upper CI'] = model.conf_int()[1]
    coefficients[borough] = coef_df
```

```python
# Print the coefficients and confidence intervals for each borough
for borough, coef_df in coefficients.items():
    print(f'Borough: {borough}')
    print(coef_df)
    print('\n')
```

```python
# Extract the slope (coefficient) for "gross_square_feet" for each borough
slopes = {}
for borough, model in linear_models.items():
    slope = model.params['gross_square_feet']
    slopes[borough] = slope

# Create a DataFrame with the slopes
NYC_slope = pd.DataFrame({'borough': list(slopes.keys()), 'slope': list(slopes.values())})

# Sort the DataFrame by slope
NYC_slope = NYC_slope.sort_values(by='slope')

# Print the resulting DataFrame
print(NYC_slope)
```

We've arranged the results in ascending order by the slope estimate. For each of the five boroughs, the t-statistic and p-value indicate that there is a relationship between sale_price and gross_square_feet. In Staten Island, an increase in square footage by one unit is estimated to increase the sale price by about 288 (USD), on average. In contrast, an increase in total square footage by one unit is estimated to result in an increase in sale price of about 4,728 (USD), on average.

# Conclusion

Our analysis showed that, in general, the gross_square_feet variable is useful for explaining, or estimating, sale_price for condominiums in New York City. We observed that removing multi-unit sales from the dataset increased model accuracy. With linear models generated for New York City as a whole, and with linear models generated for each borough individually, we observed in all cases that the t-statistic was high enough, and the p-value was low enough, to declare that there is a relationship between gross_square_feet and sale_price.

For the linear models that we generated for each individual borough, we observed a wide range in slope estimates. The slope estimate for Manhattan was much higher than the estimate for any of the other boroughs.

Finally, regression summary statistics indicate that gross_square_feet is a better single predictor of sale_price in some boroughs versus others. For example, the R-squared value was estimated at approximately 0.63 in Manhattan, and 0.59 in Bronx, compared to an estimate of only 0.35 in Queens. These differences in R-squared correspond with the scatterplots generated for each borough; the strength of sale prices versus gross square feet was higher, and the dispersion (spread), was lower for Manhattan and Brooklyn as compared to Queens where the relationship was noticeably weaker because the data was more spread out.
